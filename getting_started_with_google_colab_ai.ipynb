```python
# @title **1. Generate Text**
# Simple batch generation example
# Only text-to-text input/output is supported
from google.colab import ai

response = ai.generate_text("What is the capital of France?")
print(response)
```

```python
response = ai.generate_text("행성중에 가장 큰 행성은?")
print(response)
```

For longer text generations, you can stream the response. This displays the output token by token as it's generated, rather than waiting for the entire response to complete. This provides a more interactive and responsive experience. To enable this, simply set stream=True.

```python
# @title Simple streaming example

stream = ai.generate_text("Tell me a very short story.", stream=True)
for text in stream:
  print(text, end='')
```

```python
# @title **2. Translate languages**

stream = ai.generate_text("안녕하세요를 영어로 번역해줘", stream=True)
for text in stream:
  print(text, end='')
```

```python
stream = ai.generate_text("안녕하세요를 중국어로 번역해줘", stream=True)
for text in stream:
  print(text, end='')
```

```python
# @title **3. Write creative content**

stream = ai.generate_text("우주전쟁에 대한 짧은 글을 1000자 이내로 만들어줘", stream=True)
for text in stream:
  print(text, end='')
```

```python
# @title **4. Categorize text**
# @title Categorize text example
from google.colab import ai

# 분류할 텍스트와 카테고리를 프롬프트에 함께 입력
prompt = """
다음 텍스트를 [긍정, 부정, 중립] 중 하나로 분류해줘.

텍스트: "음식은 맛있는데 직원이 너무 불친절해서 다시는 안 갈 것 같아요."
분류:
"""

response = ai.generate_text(prompt)
print(response)
```
